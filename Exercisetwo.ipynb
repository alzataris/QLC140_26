```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

#creating a figure that is 10x5 in
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

# Coordinates for Charlotte, NC (approximate)
charlotte_lon = -80.84
charlotte_lat = 35.23

# Add an arrow saying "We are here" that is pointing to Charlotte North Carolina
ax.annotate('We are here!', xy=(charlotte_lon, charlotte_lat),
            xytext=(charlotte_lon + 10, charlotte_lat + 10), # Adjust text position slightly
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='right',
            transform=ccrs.PlateCarree())

plt.title('Map of the Earth with Charlotte, NC') # Updated title for clarity
plt.show()
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature # New import
import cartopy.mpl.ticker as mticker # New import

#creating a figure that is 10x5 in
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cfeature.OCEAN, zorder=0, color='lightblue') # Add ocean feature

# Add gridlines and labels
gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, color='gray', alpha=0.5, linestyle='--')
gl.xformatter = mticker.LongitudeFormatter() # Keeping this for potential future use or if it doesn't break
gl.yformatter = mticker.LatitudeFormatter()
gl.top_labels = False # Do not show labels on the top
gl.right_labels = False # Do not show labels on the right
gl.x_labels = False # Disable longitude labels
gl.xlines = False # Disable longitude lines

# Coordinates for Charlotte, NC (approximate)
charlotte_lon = -80.84
charlotte_lat = 35.23

# Add an arrow saying "We are here" that is pointing to Charlotte North Carolina
ax.annotate('We are here!', xy=(charlotte_lon, charlotte_lat),
            xytext=(charlotte_lon + 10, charlotte_lat + 10), # Adjust text position slightly
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='right',
            transform=ccrs.PlateCarree())

# Add major ocean labels manually
ax.text(-30, 0, 'Atlantic Ocean', transform=ccrs.PlateCarree(), horizontalalignment='center', verticalalignment='center', fontsize=12, color='darkblue', weight='bold')
ax.text(170, 0, 'Pacific Ocean', transform=ccrs.PlateCarree(), horizontalalignment='center', verticalalignment='center', fontsize=12, color='darkblue', weight='bold')
ax.text(80, -20, 'Indian Ocean', transform=ccrs.PlateCarree(), horizontalalignment='center', verticalalignment='center', fontsize=12, color='darkblue', weight='bold')

plt.title('Map of the Earth with Charlotte, NC, Oceans, Latitude Gridlines and Ocean Labels') # Updated title
plt.show()
```
